# *Feature Scaling*

# *Hierarchical Clustering*

```python
from scipy.cluster import hierarchy
```

```python
# create dendrogram
fig = plt.figure(figsize=(15,8))
dendrogram = hierarchy.dendrogram(hierarchy.linkage(df1, method="ward"))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=4)
df2 = hc.fit_predict(df1)
```

```python
df['Cluster'] = df2
```

# *KMeans*

```python
from sklearn.cluster import KMeans
```

## Elbow Graph

```python
wcss = []

for i in range(1,15):
    km = KMeans(n_clusters=i)
    km.fit_predict(df1)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,15), wcss)
plt.xlabel('K values')
plt.ylabel('WCSS')
```

```python
#k = 4
km=KMeans(n_clusters=4,random_state=0)
y1=km.fit_predict(df1)
y1
```

```python
km.cluster_centers_
```

```python
df['Cluster'] = y1
```

```python
df
```

```python
df['Cluster'].value_counts()
```

```python
df.groupby('Cluster').agg(['mean'])
```

# *DBSCAN*

```python
from sklearn.cluster import DBSCAN
```

```python
from sklearn.metrics import silhouette_score
```

```python
dbscan = DBSCAN(eps=0.98, min_samples=3)
dbscan.fit(df1)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
# Adding clusters to dataset
df['clusters']=dbscan.labels_
df
```

```python
c2 = pd.DataFrame(dbscan.labels_,columns=['Cluster ID'])
c2.value_counts()
```

```python
# Plot Clusters to dataset
plt.figure(figsize=(10, 7))  
plt.scatter(df['clusters'],df['UrbanPop'], c=dbscan.labels_) 
```
